# 考点2:数据特征提取（20分）


# a.使用corr()方法得到斯皮尔曼相关系数矩阵和输出。

```python
df_corr=df.corr()
df_corr
```

# 删除关联度大于0.8的属性。

```python
list_drop=[]
for x in df_corr.columns:
    for y in df_corr.index:
        c=df_corr.loc[x,y]
        if c>0.8 and x!=y:
            list_drop.append(x)

for x in list_drop:
    del df[x] # 删除掉大于0.8的属性
```

# 使用Heatmap绘制相关矩阵热图。

```python
import matplotlib.pyplot as plt
import seaborn as sn
plt.figure(figsize=(14,14))
sn.heatmap(df_corr)
```

# d.对于数值变量` House_State `、` Year_Income `、` Couple_Year_Income `、’填写L12_Month_Pay_Amount '中缺失的值。

```python
# 首先我们查看缺失情况
for x in df.columns:
    dfqs=df[x].isnull().sum()
    dftt=df[x].isnull().sum()/df.shape[0]*100
    print(x,'缺失值的缺失率：',dftt,'%')
```

```python
# 缺失值处理
# df['House_State']=df['House_State'].replace('ae',df['House_State'].mode()[0])
# df['Year_Income']=df['Year_Income'].replace('Be',df['Year_Income'].mode()[0])
# df['Couple_Year_Income']=df['Couple_Year_Income'].replace('Da',df['Couple_Year_Income'].mode()[0])
# df['L12_Month_Pay_Amount']=df['L12_Month_Pay_Amount'].replace('Be',df['L12_Month_Pay_Amount'].mode()[0])

# 获取每列的众数
modes = df.mode().iloc[0]

for col in ['House_State', 'Year_Income', 'Couple_Year_Income', 'L12_Month_Pay_Amount']:
    # 使用对应列的众数填充该列的NaN值
    df[col].fillna(modes[col], inplace=True)

df['Unit_Kind']=df['Unit_Kind'].replace('ae',df['Unit_Kind'].mode()[0])
df['Unit_Kind']=df['Unit_Kind'].replace('Be',df['Unit_Kind'].mode()[0])
df['Unit_Kind']=df['Unit_Kind'].replace('Da',df['Unit_Kind'].mode()[0])
df['Occupation']=df['Occupation'].replace('Y0000',df['Occupation'].mode()[0])
df['Occupation']=df['Occupation'].replace('X0000',df['Occupation'].mode()[0])
df['Occupation']=df['Occupation'].replace(' ',df['Occupation'].mode()[0])
df['Occupation']=df['Occupation'].replace('   ',df['Occupation'].mode()[0])

# 使用众数对 Nation列进行填充
df['Nation'] = df['Nation'].replace(np.nan,df['Nation'].mode()[0])
```

# e.对于名义型变量(字符串数据，DataFrame表示对象)` Gender `来填充缺失的值。

```python
# 计算'Gender'列的众数
gender_mode = df['Gender'].mode().iloc[0]

# 使用'Gender'列的众数填充该列中的缺失值（NaN）
df['Gender'].fillna(gender_mode, inplace=True)
```

```python
df.info()
```

# f.对名义变量进行独特的热编码，增强表达能力。

```python
# 转换数据
from sklearn.impute import SimpleImputer
model=SimpleImputer(missing_values=np.nan,strategy='mean')
df_model=model.fit_transform(df)

df_mean=pd.DataFrame(df_model,columns=df.columns)
# 做独热编码
df_one=pd.get_dummies(df_mean)#做独热编码
```

```python
df_one
```

# 考点 3: 数据集分割（20分）

## a.根据目标列和其他列拆分数据，并将它们命名为X和Y。

```python
# 取出X
x=[col for col in df_one.columns if col!='Cust_No' and col!='Target']
x
```

```python
X=df_one[x]
X
```

```python
# 取出Y
Y=df_one['Target']
Y
# X 数据进行独热
```

## b.使用sklearn.train_test_split将数据X和Y划分为训练集和测试集，划分测试集的比例为0.3。

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.3)
```

## c.引入随机森林算法随机指定参数。

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=42)
```

# 任务2:建立网络
考点1: 建立网络（40分
## a.模型训练。
## b.使用准确率对模型进行评分。

```python
# 训练使用准确率对模型进行评分。
df_rf = rf.fit(x_train,y_train)
# 准确率评分
print(df_rf.score(x_train,y_train))
print(df_rf.score(x_test,y_test))
```

## c.引入sklearn.SMOTE过采样率=0.1。


```python
# 引入sklearn.SMOTE过采样率=0.1。
# 引入imblearn的SMOTE模块，设定过采样率为0.1
from imblearn.over_sampling import SMOTE

x_train,y_train=SMOTE(sampling_strategy=0.1).fit_resample(x_train,y_train)
pd.Series(y_train).value_counts()
```

```python
#再次进行随机深林模型
rf = RandomForestClassifier()
df_R=rf.fit(x_train,y_train)
rf_pred=df_R.predict(x_test)
print(df_R.score(x_test,y_test))
```

# 同时使用XGBOST、GBDT等算法分别建模后进行模型比对。

```python
#使用XGBoost
xgb_model = xgb.XGBClassifier(n_estimators=100, max_depth=5, random_state=42)
xgb_model.fit(x_train, y_train)
xgb_pred = xgb_model.predict(x_test)
print("XGBoost结果：")
print(accuracy_score(y_test, xgb_pred))

#使用GBDT
from sklearn.ensemble import GradientBoostingClassifier
gbdt_model = GradientBoostingClassifier(n_estimators=100, max_depth=5, random_state=42)
gbdt_model.fit(x_train, y_train)
gbdt_pred = gbdt_model.predict(x_test)
print("GBDT结果：")
print(accuracy_score(y_test, gbdt_pred))

# #使用SVM
# from sklearn.svm import SVC
# svm_model = SVC(kernel='linear', C=1, random_state=42)
# svm_model.fit(x_train, y_train)
# svm_pred = svm_model.predict(x_test)
# print("SVM结果：")
# print(accuracy_score(y_test, svm_pred))
```

## 使用f1_score评估上面三种模型好坏。


```python
# 使用f1_score评估上面三种模型好坏，随机森林，XGBoost，GBDT，SVM。
from sklearn.metrics import f1_score
print("随机森林结果：")
rf_score = f1_score(y_test, rf_pred)
print(f1_score(y_test, rf_pred))
print("XGBoost结果：")
xgb_score = f1_score(y_test, xgb_pred)
print(f1_score(y_test, xgb_pred))
print("GBDT结果：")
gbdt_score = f1_score(y_test, gbdt_pred)
print(f1_score(y_test, gbdt_pred))

```
