# 三. 图形绘制流程
1. 创建画布：plt.figure()
         plt.figure(figsize=(), dpi=) 返回一个fig对象
         其中第一个参数是指定图的长宽，第二个指定图形的清晰度
2. 绘制图像：plt.plot(x,y) x：x轴，y:y轴
3. 显示图像：plt.show() **注意**：保存图片的时候，一定要在show之前保存，
         不然会得到一张空白图，因为在show()之后，会自动调用clear()清除缓存

# 四.绘图测试

````python
import matplotlib.pyplot as plt

plt.figure(figsize=(5,10), dpi=100)

xarrays = [i for i in range(10)]
yarrays = [i**2 for i in range(10)]

plt.plot(xarrays,yarrays)

plt.show()


'''
绘画正弦函数和余弦函数

'''

import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0,4*np.pi,0.1)   # start,stop,step
y = np.sin(x)
z = np.cos(x)
plt.plot(x,y,x,z)
plt.show()
````

## 四.一 添加刻度
1. 为什么要添加刻度呢？观察上图，我们发现上面的x轴和y轴都是只显示偶数，其它的奇数未显示，这样在展示实验效果或放入文章中都会影响其可读性。因此我们要用matplotlib的函数xticks()和yticks()来对x轴和y轴进行刻度调整。

````python
# 添加刻度

import numpy as np
import matplotlib.pyplot as plt
x = range(1,13,1)
y = range(1,13,1)
plt.plot(x,y)
print("添加刻度线前:")
plt.show()


plt.plot(x,y)
plt.xticks(x) # 下标从0到12
plt.yticks(y)
print("添加刻度线后：")

plt.show()


# 修改刻度名称
# 如把x轴的1~12修改为月份的名称
import calendar

print("修改名称后：")
plt.plot(x,y)
plt.xticks(x, calendar.month_name[1:13],color='green',rotation=300)
plt.show()

#添加标题和标签名
plt.title("month of year")
plt.xlabel("month")
plt.ylabel("ddd")
plt.plot(x,y)
plt.xticks(x, calendar.month_name[1:13],color='green',rotation=300)
plt.show()

````

# 添加网格线
可以用grid()方法来添加网格线
```cpp
matplotlib.pyplot.grid(b=None, which='major', axis='both', )
    b：可选，默认为 None，可以设置布尔值，true 为显示网格线，false 为不显示，如果设置 **kwargs 参数，则值     为 true。
    which：可选，可选值有 'major'、'minor' 和 'both'，默认为 'major'，表示应用更改的网格线。
    axis：可选，设置显示哪个方向的网格线，可以是取 'both'（默认），'x' 或 'y'，分别表示两个方向，x 轴方向     或 y 轴方向。
    **kwargs：可选，设置网格样式，可以是 color='r', linestyle='-' 和 linewidth=2，分别表示网格线的颜色，     样式和宽度。
    
```


````python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 2, 3, 4])
y = np.array([1, 4, 9, 16])


plt.title("my_grid()_Test")
plt.xlabel("x_label")
plt.ylabel("y_label")

plt.plot(x, y)

plt.grid()

plt.show()

# 修改网格线的类型
plt.plot(x,y)
plt.grid(color='b', linestyle='--')
x = range(1,17, 1)
plt.yticks(x)
plt.show()
````
