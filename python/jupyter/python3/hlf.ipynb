```python
# 画出饼状图
import matplotlib.pyplot as plt

# 准备数据
labels = ['你好', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']

# 创建饼状图
plt.figure(figsize=(8, 6))
plt.pie(sizes, 
        labels=labels, 
        colors=colors,
        autopct='%1.1f%%',  # 显示百分比
        startangle=90,      # 起始角度
        explode=(0.1, 0, 0, 0))  # 突出显示第一块

# 添加标题并保持长宽比相等
plt.title("Sample Pie Chart", fontsize=16)
plt.axis('equal')

# 显示图表（在Jupyter中会自动渲染）
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib

# 设置中文环境和解决负号显示
matplotlib.rcParams['font.sans-serif'] = ['SimHei']
matplotlib.rcParams['axes.unicode_minus'] = False

# 数据准备
labels = ['18岁以下', '18-25岁', '26-35岁', '36-45岁', '46岁以上']
sizes = [4, 94, 1, 1, 1]
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0']

# 创建饼图
plt.figure(figsize=(5, 4))
wedges, texts, autotexts = plt.pie(
    sizes,
    colors=colors,
    autopct='%1.1f%%',
    startangle=90,
    pctdistance=0.9,  # 调整百分比位置
    explode=(0.1, 0.3, 0.3, 0.3, 0.3)
)

# 添加图例并调整位置
plt.legend(
    wedges,
    labels,
    title="年龄分布",
    loc='center left',
    bbox_to_anchor=(1, 0, 0.5, 1)  # 将图例放置在图表右侧
)

# 设置标题和等轴比例
plt.title("题目1：您的年龄分布", fontsize=16)
plt.axis('equal')

# 自动调整布局
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib

# 设置中文环境和解决负号显示
matplotlib.rcParams['font.sans-serif'] = ['SimHei']
matplotlib.rcParams['axes.unicode_minus'] = False

# 数据准备
labels = ['18岁以下', '18-25岁', '26-35岁', '36-45岁', '46岁以上']
sizes = [4, 94, 1, 1, 1]
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0']

# 创建饼图
plt.figure(figsize=(10, 8))
wedges, texts, autotexts = plt.pie(
    sizes,
    colors=colors,
    autopct='%1.1f%%',
    startangle=90,
    pctdistance=0.85,  # 调整百分比标签与圆心距离
    labeldistance=1.05,  # 调整分类标签与圆心距离
    explode=(0.1, 0, 0, 0, 0),
    textprops={'fontsize': 12}  # 设置文本属性
)

# 调整百分比标签位置到外侧
for autotext in autotexts:
    autotext.set_color('black')  # 设置标签颜色
    autotext.set_fontsize(12)    # 设置标签字号
    # 将标签位置外移
    x, y = autotext.get_position()
    autotext.set_position((x * 1.15, y * 1.15))  # 1.15为外移系数，可调整

# 添加图例
plt.legend(
    wedges,
    labels,
    title="年龄分布",
    loc='center left',
    bbox_to_anchor=(1, 0.5)
)
plt.xticks(rotation=-15) 
# 设置标题
plt.title("题目1：您的年龄分布", fontsize=16)
plt.axis('equal')

# 自动调整布局
plt.tight_layout()
plt.show()
```
